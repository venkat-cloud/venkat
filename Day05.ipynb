## idiomatic python
- transformation of code into beautiful idiomatic python
- replaces traditional index manipulation with python looping idiioms

```python
# looping over range of numbers
for i in [0,1,2,3,4,5]:
    print(i**2,end=' ')
# pythonic way--
print()
for i in range(6):
    print(i**2,end=' ')
```

```python
#looping over the colection(list)
li=[1,2,3,4,5]
for i in range(len(li)):
    print(li[i],end=' ' )
print()
# pythonic way---
for i in li:
    print(i,end=' ')
```

```python
# looping from backwards
li=[1,2,3,4,5]
for i in range(len(li)-1,-1,-1):
    print(li[i],end=' ')
print()
#pythonic way--
for i in reversed(li):
    print(i,end=' ')
```

```python
#looping over collection with index
for i in range(len(li)):
    print(i,'-->',li[i])
print()
#pythonic way
for i,a in enumerate(li):
    print(i,'-->',a)
```

```python
#looping over two collections
l1=[1,2,34,5,6,5]
l2=[1,23,4,5,6,7,8,9,0]
n=min(len(l1),len(l2))
for i in range(n):
    print(l1[i],' ',l2[i])
#pythonic way
print()
for i,a in zip(l1,l2):
    print(i,' ',a)
    
```

```python
# loopimg in sorted order
li=[3,5,21,4,1,8,0,5]
for i in sorted(li):
    print(i,end=' ')
print()
for i in reversed(sorted(li)):
    print(i,end=' ')
```

```python
#traditional approach
li=[]
for i in range(10):
    a=i**2
    li.append(a)
print(sum(li))
#pythonic way---
print(sum(i**2 for i in range(10)))
```
